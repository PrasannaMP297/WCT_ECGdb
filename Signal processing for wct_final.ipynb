```python
def process_signal(data):
    list_string_signal = data
    if list_string_signal[0]=='':
        list_string_signal.remove('')
    signal = []
    for s in list_string_signal:
        z = s.translate({ord('e'): None})
        signal.append(z)

    signal_data_split=[]
    for one_signal in signal:
        part_signal=one_signal.split('.')
    if len(part_signal)>2:
        for i in range(1,len(part_signal)):
            signal_data_split.append(int(part_signal[i])/(10**len(part_signal[i])))
    else:
        signal_data_split.append(int(part_signal[-1])/(10**len(part_signal[-1])))

    signal_data_split=np.array(signal_data_split)
    return signal_data_split
    
```

```python
# signal = 
def signal_process(signal):
    j=0
    real= []
    imag = []
    while j<=91:
        list_string_signal=signal
        if list_string_signal[0]=='':
            list_string_signal.remove('')
        signal = []
        for s in list_string_signal:
            z = s.translate({ord('e'): None})
            signal.append(z)

        signal_data_split=[]
        for one_signal in signal:
            part_signal=one_signal.split('.')
            if len(part_signal)>2:
                for i in range(1,len(part_signal)):
                    signal_data_split.append(int(part_signal[i])/(10**len(part_signal[i])))
            else:
                signal_data_split.append(int(part_signal[-1])/(10**len(part_signal[-1])))

        signal_data_split=np.array(signal_data_split)
    #     import matplotlib.pyplot as plt
    #     plt.plot(signal_data_split)
    #     plt.show()
        sp = np.fft.fft(signal_data_split)
        freq = np.fft.fftfreq(signal_data_split.shape[-1])
    #     plt.plot(freq, sp.real)
    #     plt.show()
        real.append(sp.real)
        imag.append(sp.imag)

        j+=1
        return real,imag
```
